```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Supondo que 'data' é o DataFrame que contém seus dados
# e 'y_column' é o nome da coluna que você deseja codificar
y_column = 'Function'

# Criar uma instância do LabelEncoder
label_encoder = LabelEncoder()

# Ajustar e transformar os dados da coluna alvo usando LabelEncoder
data_features[y_column] = label_encoder.fit_transform(data_features[y_column])

# Verificar os valores únicos após a codificação
print(data_features[y_column].unique())
```

```python
data_features
```

```python
y = data_features.iloc[:,2]
y
```

```python
X = data_features.iloc[:,3:]  #independent columns
y = data_features.iloc[:,2]    #target column Function
```

Univariate Selection

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2


#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

Feature Importance

```python

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt

model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization


feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='pie') #barh ou bar


plt.title("Feature Importance")
plt.show()
```

```python
# Calculate string lengths for each label
#data_features['length'] = data_features['Sequence'].apply(len)

# Split the DataFrame by label
label_0_lengths = data_features[data_features['Function'] == 0]['length']
label_1_lengths = data_features[data_features['Function'] == 1]['length']

# Create histograms
plt.hist(label_0_lengths, alpha=0.5, label='non-AMP', bins=10, color="cyan")
plt.hist(label_1_lengths, alpha=0.5, label='AMP', bins=10, color ="blue")

plt.xlabel('String Length')
plt.ylabel('Frequency')
plt.title('Distribution of String Lengths by Function')
plt.legend()
plt.yscale('log')
plt.show()
```

```python
# Split the DataFrame by label
label_0_lengths = data_features[data_features['Function'] == 0]['MW_modlamp']
label_1_lengths = data_features[data_features['Function'] == 1]['MW_modlamp']

# Create histograms
plt.hist(label_0_lengths, alpha=0.5, label='non-AMP', bins=10, color= "cyan")
plt.hist(label_1_lengths, alpha=0.5, label='AMP', bins=10, color= "blue")

plt.xlabel('Molecular Weight (g/mol)')
plt.ylabel('Frequency')
plt.title('Distribution of Molecular Weight by Function')
plt.legend()
plt.yscale('log')
plt.show()
```

```python
# Select relevant columns
data_features['Function'] = data_features['Function'].map({0: 'AMP', 1: 'non-AMP'})

df_selected = data_features[['Function', 'A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']]

# Reshape the DataFrame to long format
df_long = pd.melt(df_selected, id_vars=['Function'], var_name='Amino Acid', value_name='Count')

# Display the reshaped DataFrame
print(df_long.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a grouped bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='Amino Acid', y='Count', hue='Function', data=df_long, palette='muted')

# Add titles and labels
plt.title('Distribution of Amino Acids in AMP vs Non-AMP')
plt.xlabel('Amino Acid')
plt.ylabel('Count')

# Display the plot
plt.show()
```
