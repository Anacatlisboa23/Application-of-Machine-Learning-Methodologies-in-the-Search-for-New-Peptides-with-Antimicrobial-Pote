# PCA:

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv('/content/final_com_featuress (1).csv')

# Supondo que as colunas 'alvo' e 'sequencia' são as que você quer manter
target_column = 'Function'
sequence_column = 'Sequence'

# Separar as características, o alvo e as sequências
X = df.drop([target_column, sequence_column], axis=1)
y = df[target_column]
sequences = df[sequence_column]

# Codificar o alvo (y) se necessário
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Padronizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Executar PCA
pca = PCA(n_components=None)
X_pca = pca.fit_transform(X_scaled)

# Analisar a variação explicada por cada componente principal
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='--')
plt.title('Cumulative Variance Explained by each Principal Component')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()

# Coeficientes dos componentes principais
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(X_scaled.shape[1])], index=X.columns)

# Identificar as features mais importantes
num_pc = 3  # número de componentes principais que você quer examinar
important_features = set()

for i in range(num_pc):
    important_features.update(loadings.iloc[:, i].abs().sort_values(ascending=False).head(10).index)

print(f'Features mais importantes: {important_features}')

# Criar um novo DataFrame com as features mais importantes
X_important = X[list(important_features)]

# Adicionar as colunas alvo e sequências
df_important = X_important.copy()
df_important[target_column] = y
df_important[sequence_column] = sequences

# Salvar o novo DataFrame
df_important.to_csv('features_mais_importantes.csv', index=False)

print('DataFrame com as features mais importantes salvo em "features_mais_importantes.csv"')



```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA

# Carregar os dados
df = pd.read_csv('/content/final_com_featuress (1).csv')

# Definir as colunas alvo e sequência
target_column = 'Function'
sequence_column = 'Sequence'

# Separar características, alvo e sequência
X = df.drop([target_column, sequence_column], axis=1)
y = df[target_column]
sequences = df[sequence_column]

# Codificar o alvo
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Padronizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Executar PCA
pca = PCA(n_components=None)
X_pca = pca.fit_transform(X_scaled)

# Coeficientes dos componentes principais
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(X_scaled.shape[1])], index=X.columns)

# Identificar as features mais importantes
num_pc = 3  # número de componentes principais que você quer examinar
important_features = set()

for i in range(num_pc):
    important_features.update(loadings.iloc[:, i].abs().sort_values(ascending=False).head(10).index)

print(f'Features mais importantes: {important_features}')

# Criar um novo DataFrame com as features mais importantes
X_important = X[list(important_features)]

# Adicionar as colunas alvo e sequência
df_important = X_important.copy()
df_important[target_column] = y
df_important[sequence_column] = sequences

# Salvar o novo DataFrame
df_important.to_csv('features_mais_importantes.csv', index=False)

print('DataFrame com as features mais importantes salvo em "features_mais_importantes.csv"')

# Plotar as importâncias das features
# Coletar as importâncias das features para as 3 principais componentes principais
feature_importance_df = loadings.loc[list(important_features)].abs().mean(axis=1).sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=feature_importance_df.values, y=feature_importance_df.index, palette='viridis')
plt.title('Feature importance using Lasso Regression')
plt.xlabel('Coefficients')
plt.ylabel('Features')
plt.grid(axis='x')
plt.show()
```
