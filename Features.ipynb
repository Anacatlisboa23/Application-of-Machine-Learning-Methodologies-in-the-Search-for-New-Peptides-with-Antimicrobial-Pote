Univariate Selection

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2


#apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X_scaled,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

Feature Importance

```python

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt

model = ExtraTreesClassifier()
model.fit(X_scaled,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization


feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='pie') #barh ou bar


plt.title("Feature Importance")
plt.show()
```

# PCA, vamos manter as caracteristicas importantes

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

```

```python
# Separar features e target
X = data_features.drop(columns=['Function',"Sequence"])  # Substitua 'target' pelo nome da coluna alvo
y = data_features['Function']  # Substitua 'target' pelo nome da coluna alvo
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X
```

```python
# Treinar um modelo de Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_scaled, y)

# Ajustar a seleção de features com um threshold mais baixo
selector = SelectFromModel(clf, threshold="mean", prefit=True)
X_important = selector.transform(X_scaled)

# Verificar se alguma feature foi selecionada, se não, usar um threshold mais baixo
if X_important.shape[1] == 0:
    selector = SelectFromModel(clf, threshold="median", prefit=True)
    X_important = selector.transform(X_scaled)

# Obter as colunas selecionadas pelo seletor
selected_columns = X.columns[selector.get_support()]

# Criar um novo DataFrame apenas com as features importantes
data_important = pd.DataFrame(X_important, columns=selected_columns)

# Adicionar colunas 'Function' e 'Sequence' de volta no início do DataFrame
data_important.insert(0, 'Sequence', data_features['Sequence'])  # 'Sequence' como primeira coluna
data_important.insert(1, 'Function', y.values)  # 'Function' como segunda coluna

# Salvar o novo dataset
data_important.to_csv('dataset_importante.csv', index=False)

print("Features importantes selecionadas:", selected_columns)
```
