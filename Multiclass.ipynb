```python
SS_classifier = svm.SVC(kernel='linear', C=0.1, gamma=0.1, coef0=0.0, degree=2)  # Hiperparâmetros
scores_ss = []
f1s_ss = []
recls_ss = []
precisions_ss = []
mccs_ss = []

for train_index, test_index in s.split(X_tr, y_tr):
    X_SS_train, X_SS_test = X_tr[train_index, 4:], X_tr[test_index, 4:]
    y_SS_train, y_SS_test = y_tr[train_index], y_tr[test_index]

    X_SS_train = min_max_scaler.fit_transform(X_SS_train)
    X_SS_test = min_max_scaler.transform(X_SS_test)

    SS_classifier.fit(X_SS_train, y_SS_train)
    ypred = SS_classifier.predict(X_SS_test)

    # Calcular as métricas
    f1 = f1_score(y_SS_test, ypred, average='weighted', zero_division=1)
    recall = recall_score(y_SS_test, ypred, average='weighted', zero_division=1)
    precision = precision_score(y_SS_test, ypred, average='weighted', zero_division=1)
    mcc = matthews_corrcoef(y_SS_test, ypred)

    scores_ss.append(SS_classifier.score(X_SS_test, y_SS_test))
    f1s_ss.append(f1)
    recls_ss.append(recall)
    precisions_ss.append(precision)
    mccs_ss.append(mcc)

# Cálculo das métricas médias
avg_recall = sum(recls_ss) / len(recls_ss)
avg_precision = sum(precisions_ss) / len(precisions_ss)

print("----------------------------------------")
print("Scores: ", np.min(scores_ss), np.max(scores_ss), np.std(scores_ss))
print("F1s: ", np.min(f1s_ss), np.max(f1s_ss), np.std(f1s_ss))
print("MCCs: ", np.min(mccs_ss), np.max(mccs_ss), np.std(mccs_ss))
print("Recalls: ", np.min(recls_ss), np.max(recls_ss), np.std(recls_ss))
print("Precisions: ", np.min(precisions_ss), np.max(precisions_ss), np.std(precisions_ss))
print("avg cross-validation accuracy:", (sum(scores_ss) / len(scores_ss)))
print("avg cross-validation f1:", (sum(f1s_ss) / len(f1s_ss)))
print("avg cross-validation mcc:", (sum(mccs_ss) / len(mccs_ss)))
print("avg cross-validation recall:", avg_recall)
print("avg cross-validation precision:", avg_precision)
```

```python
from sklearn.metrics import (
    f1_score, recall_score, precision_score, matthews_corrcoef, confusion_matrix, accuracy_score
)
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC

# Selecionando as colunas desejadas e escalando os dados
X_new = X_tr[:, 4:]
min_max_scaler = MinMaxScaler()
X_new = min_max_scaler.fit_transform(X_new)

# Treinando o classificador SVM
SVM_classifier = SVC(random_state=42)
SVM_classifier.fit(X_new, y_tr)

# Fazendo previsões no conjunto de treino
y_tr_predict = SVM_classifier.predict(X_new)

# Calculando as métricas de avaliação
f1 = f1_score(y_tr, y_tr_predict, average='macro')
mcc = matthews_corrcoef(y_tr, y_tr_predict)
recall = recall_score(y_tr, y_tr_predict, average='macro')
precision = precision_score(y_tr, y_tr_predict, average='macro')
accuracy = accuracy_score(y_tr, y_tr_predict)

print('f1 on Train set: ', f1)
print('MCC on Train set: ', mcc)
print('Accuracy on Train set: ', accuracy)
print('Recall on Train set: ', recall)
print('Precision on Train set: ', precision)

# Calculando a matriz de confusão e métricas adicionais
conf_matrix = confusion_matrix(y_tr, y_tr_predict)
print('Confusion Matrix:')
print(conf_matrix)

# Para especificidade e sensibilidade em multiclasse, você pode calcular para cada classe
# e depois tirar a média
specificities = []
sensitivities = []
for i in range(len(conf_matrix)):
    tn = np.sum(np.delete(np.delete(conf_matrix, i, 0), i, 1))
    fp = np.sum(np.delete(conf_matrix, i, 0)[:, i])
    fn = np.sum(np.delete(conf_matrix, i, 1)[i, :])
    tp = conf_matrix[i, i]
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 0
    specificities.append(specificity)
    sensitivities.append(sensitivity)

print('Specificity on Train set (average): ', np.mean(specificities))
print('Sensitivity on Train set (average): ', np.mean(sensitivities))

```

nota: Naive Bayes Multinomial, ele não suporta a saída de probabilidades necessárias para calcular a Área sob a Curva ROC (AUROC) diretamente. Isso ocorre porque o Naive Bayes Multinomial é um classificador probabilístico baseado em contagem, e não retorna probabilidades diretas para cada classe.

Uma maneira alternativa de avaliar a performance do seu modelo Naive Bayes Multinomial é utilizando a curva ROC e a AUROC em combinação com um classificador que suporte a saída de probabilidades.
