# Megares E LAMP FASTA TO CSV

```python
from Bio import SeqIO
import pandas as pd

# Função para extrair informações de sequências Fasta
def extract_sequence_info(record):
    seq_id = record.id
    description = record.description
    sequence = str(record.seq)

    organism = description.split('|')[1].strip() if '|' in description else None
    activity = description.split('|')[-1].strip() if '|' in description else None

    return {'ID': seq_id, 'Organism': organism, 'Activity': activity, 'Sequence': sequence}

# Função para converter dados Fasta para DataFrame
def fasta_to_dataframe(file_path):
    records = SeqIO.parse(file_path, 'fasta')
    data = [extract_sequence_info(record) for record in records]
    return pd.DataFrame(data)

# Ler os dados dos arquivos Fasta
fasta_file1 = r'C:\Users\pc\Desktop\Dataset\LAMP\LAMP2.FASTA'
fasta_file2 = r'C:\Users\pc\Desktop\Dataset\MEGARes\megares_database_v3.00.fasta'

df1 = fasta_to_dataframe(fasta_file1)
df2 = fasta_to_dataframe(fasta_file2)

# Salvar os DataFrames em arquivos CSV
csv_file1 = 'LAMP.csv'
csv_file2 = 'MEGARES.csv'

df1.to_csv(csv_file1, index=False)
df2.to_csv(csv_file2, index=False)

print(f'Arquivos CSV foram criados: {csv_file1} e {csv_file2}')
```

__DRAMP__

```python
import os
import csv
```

```python
#Passar de arquivo Fasta a arquivo CSV:

def fasta_to_csv(input_fasta, output_csv):
    sequences = []

    with open(input_fasta, 'r') as fasta_file:
        sequence_id = None
        sequence = ""

        for line in fasta_file:
            line = line.strip()

            if line.startswith('>'):
                if sequence_id is not None:
                    sequences.append([sequence_id, sequence])
                sequence_id = line[1:]
                sequence = ""
            else:
                sequence += line

        if sequence_id is not None:
            sequences.append([sequence_id, sequence])

    with open(output_csv, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file, delimiter=';')
        csv_writer.writerow(['Sequence ID', 'Sequence'])

        for seq_id, seq in sequences:
            csv_writer.writerow([seq_id, seq])

def process_fasta_files(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith(".fasta"):
            input_fasta_file = os.path.join(input_folder, filename)
            output_csv_file = os.path.join(output_folder, filename.replace(".fasta", ".csv"))
            fasta_to_csv(input_fasta_file, output_csv_file)

if __name__ == "__main__":
    input_folder = 'C:\\Users\\pc\\Desktop\\Dataset\\DRAMP'  # Caminho para a pasta que contém os arquivos .fasta
    output_folder = 'C:\\Users\\pc\\Desktop\\Dataset\\DRAMP\\output_folder'

    process_fasta_files(input_folder, output_folder)


```

```python
if response.status_code == 200:
    # A resposta foi bem-sucedida (código 200)
    dados = response.json()  # Se a resposta estiver em formato JSON

    # Imprimir os dados para verificar
    print(dados)

    # Salvar os resultados em um arquivo Excel (XLSX)
    nome_do_arquivo = 'resultados_api.xlsx'
    with pd.ExcelWriter(nome_do_arquivo, engine='openpyxl') as writer:
        # Verificar se 'data' está presente nos dados
        if 'data' in dados:
            resultados = dados['data']
        else:
            resultados = dados  # Se 'data' não estiver presente

        pd.DataFrame(resultados).to_excel(writer, index=False)

    print(f"Resultados salvos em {nome_do_arquivo}")
else:
    # A resposta teve um código de erro
    print(f"Erro na solicitação: {response.status_code}")
    print(response.text)  # Imprime detalhes da resposta em caso de erro
```

Palavra Chave  Uniprot  : '_Defensis_' | '_Quorum sensing_'


```python
import csv

def parse_header(header):
    parts = header.split('|')
    sequence_id = parts[1]
    protein_name = parts[2].split(' ')[0]
    description_parts = parts[2].split(' ')[1:]

    species = ''
    gene_name = ''

    for part in description_parts:
        if part.startswith('OS='):
            species = part[3:]
        elif part.startswith('GN='):
            gene_name = part[3:]

    return sequence_id, protein_name, ' '.join(description_parts), species, gene_name

def fasta_to_csv(input_fasta, output_csv):
    sequences = []

    with open(input_fasta, 'r') as fasta_file:
        sequence_id = None
        description = ""
        species = ""
        gene_name = ""
        sequence = ""

        for line in fasta_file:
            line = line.strip()

            if line.startswith('>'):
                if sequence_id is not None:
                    sequences.append([sequence_id, description, species, gene_name, sequence])
                sequence_id, protein_name, description, species, gene_name = parse_header(line)
                sequence = ""
            else:
                sequence += line

        # Append the last sequence
        if sequence_id is not None:
            sequences.append([sequence_id, description, species, gene_name, sequence])

    with open(output_csv, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file, delimiter=';')
        csv_writer.writerow(['Sequence ID', 'Description', 'Species', 'Gene Name', 'Sequence'])

        for seq_id, desc, spec, gn, seq in sequences:
            csv_writer.writerow([seq_id, desc, spec, gn, seq])

if __name__ == "__main__":
    # Defensis
    input_fasta_file_defensis = 'C:\\Users\\pc\\Desktop\\Dataset\\Negative cases\\defensis\\uniprotkb_defensins_2024_02_16.fasta'
    output_csv_file_defensis = 'uniprotkb_defensins3.csv'
    fasta_to_csv(input_fasta_file_defensis, output_csv_file_defensis)

    # Quorum Sensing
    input_fasta_file_quorum = 'C:\\Users\\pc\\Desktop\\Dataset\\Negative cases\\uniprotkb_quorum_sensing_2024_02_16.fasta\\uniprotkb_quorum_sensing_2024_02_16.fasta'
    output_csv_file_quorum = 'uniprotkb_quorum_sensing3.csv'
    fasta_to_csv(input_fasta_file_quorum, output_csv_file_quorum)




```
